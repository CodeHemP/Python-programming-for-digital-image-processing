# 使用pillow包进行缩放和旋转

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

im = Image.open(r'd:\myjpg\30.jpg')

# Provide the target width and height of the image
(width, height) = (im.width // 2, im.height // 2)
im_resized = im.resize((width, height))
plt.figure()
plt.imshow(im_resized)

(width, height) = (width // 2, height // 2)
im_resized = im.resize((width, height))
plt.figure()
plt.imshow(im_resized)

(width, height) = (width // 2, height // 2)
im_resized = im.resize((width, height))
plt.figure()
plt.imshow(im_resized)

# Rotate the image by 60 degrees counter clockwise
theta = 60
# Angle is in degrees counter clockwise
im_rotated = im.rotate(angle=theta)
plt.figure()
plt.imshow(im_rotated)
```

# 使用numpy对像元处理进行缩放，作业做放大

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

img_resize_array=np.zeros((img_array.shape),dtype=int)
print(img_resize_array.shape)
for k in range(img_array.shape[2]):
    for i in range(img_array.shape[0]):
        if(i%2==0):
            for j in range(img_array.shape[1]):
                if(j%2==0):
                    img_resize_array[i,j,k]=img_array[i,j,k]
plt.figure()
plt.imshow(img_resize_array)

band=img_array.shape[2]
height=math.ceil(img_array.shape[0]/2)
width=math.ceil(img_array.shape[1]/2)
print(height,width)
img_resize_array2=np.zeros((height,width,band),dtype=int)
print(img_resize_array2.shape)

for k in range(img_array.shape[2]):
    m=0
    for i in range(img_array.shape[0]):
        n=0
        if(i%2==0):
            for j in range(img_array.shape[1]):
                if(j%2==0):
                    img_resize_array2[m,n,k]=img_array[i,j,k]
                    n=n+1
            m=m+1
plt.figure()
plt.imshow(img_resize_array2)

band=img_resize_array2.shape[2]
height=math.ceil(img_resize_array2.shape[0]/2)
width=math.ceil(img_resize_array2.shape[1]/2)
print(height,width)
img_resize_array3=np.zeros((height,width,band),dtype=int)
print(img_resize_array3.shape)

for k in range(img_resize_array2.shape[2]):
    m=0
    for i in range(img_resize_array2.shape[0]):
        n=0
        if(i%2==0):
            for j in range(img_resize_array2.shape[1]):
                if(j%2==0):
                    img_resize_array3[m,n,k]=img_resize_array2[i,j,k]
                    n=n+1
            m=m+1
plt.figure()
plt.imshow(img_resize_array3)

plt.show()
```

# 使用numpy降低亮度和设置四个灰度级，作业做降低对比度

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math

img = Image.open(r'd:\myjpg\19.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

print(type(img_array[0,0,0]))

plt.figure()
plt.imshow(img_array//2)

img_resize_array=np.zeros((img_array.shape),dtype=int)
print(img_resize_array.shape)
for k in range(img_array.shape[2]):
    for i in range(img_array.shape[0]):
            for j in range(img_array.shape[1]):
                if(img_array[i,j,k]<50):
                    img_resize_array[i,j,k]=50
                elif(img_array[i,j,k]<100):
                    img_resize_array[i,j,k]=100
                elif(img_array[i,j,k]<200):
                    img_resize_array[i,j,k]=150
                else:
                    img_resize_array[i,j,k]=200
                
plt.figure()
plt.imshow(img_resize_array)

plt.show()
```
