```python
filename=r'dipum_images\dipum_images_ch03\Fig0318(b)(ckt-board-slt-pep-both-0pt2).tif'
img=Image.open(filename)
img_array=np.array(img)
print(img_array.shape)
plt.imshow(img_array,'gray')
```

```python
height=img_array.shape[0]
width=img_array.shape[1]
print(height,width)
```

```python
img_new_array=np.zeros((height,width),np.int)
print(img_new_array.shape)
```

```python
def myGetSlipWin3x3Mean(img_array,i,j):
    mean=0
    for m in range(-1,2):
        for n in range(-1,2):
            mean=mean+img_array[i+m,j+n]/9
    return round(mean)
```

```python
def myGetSlipWin3x3Middle(img_array,i,j):
    middle=[]
    for m in range(-1,2):
        for n in range(-1,2):
            middle.append(img_array[i+m,j+n])
    middle=(sorted(middle))
    return middle[4]
```

```python
def myGetSlipWin5x5Middle(img_array,i,j):
    middle=[]
    for m in range(-2,3):
        for n in range(-2,3):
            middle.append(img_array[i+m,j+n])
    middle=(sorted(middle))
    return middle[12]
```

```python
for i in range(1,height-1):
    for j in range(1,width-1):
        img_new_array[i,j]=myGetSlipWin3x3Mean(img_array,i,j)
```

```python
plt.imshow(img_new_array,'gray')
```

```python
for i in range(1,height-1):
    for j in range(1,width-1):
        img_new_array[i,j]=myGetSlipWin3x3Middle(img_array,i,j)
```

```python
plt.imshow(img_new_array,'gray')
```

```python
for i in range(1,height-1):
    for j in range(1,width-1):
        img_new_array[i,j]=myGetSlipWin3x3Middle(img_new_array,i,j)
```

```python
plt.imshow(img_new_array,'gray')
```

```python
for i in range(2,height-2):
    for j in range(2,width-2):
        img_new_array[i,j]=myGetSlipWin5x5Middle(img_array,i,j)
```
