# 增加对比度图像增强

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
img=Image.open("dipum_images/dipum_images_ch03/Fig0308(a)(pollen).tif")
plt.figure()
plt.imshow(img)
```

```python
img_array=np.array(img)
print(img_array.min(),img_array.max())

ar = img_array.flatten() 
plt.hist(ar,bins = 256,range=(1,255))

img_org = Image.fromarray(img_array)
plt.figure()
plt.imshow(img_org)
```

```python
img_new = Image.fromarray(img_array*3)
img_array=img_array*3
ar = img_array.flatten() 
plt.hist(ar,bins = 256,range=(1,255))
plt.figure()
plt.imshow(img_new)
```
