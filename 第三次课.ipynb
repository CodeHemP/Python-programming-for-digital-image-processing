# 图像任意倍数缩放

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

height=img_array.shape[0]*3
width=img_array.shape[1]*3
band=img_array.shape[2]
img_resize_array=np.zeros((height,width,band),dtype=int)
print(img_resize_array.shape)

#放大三倍
for k in range(band):
    for x in range(height):
        for y in range(width):
            i=round(x/3)
            j=round(y/3)
            if(i>=0 and i<img_array.shape[0] and j>=0 and j<img_array.shape[1]):
                img_resize_array[x,y,k]=img_array[i,j,k]      
        
# for k in range(band):
#     for i in range(img_array.shape[0]):
#         for j in range(img_array.shape[1]):
#             x=round(i*3)
#             y=round(j*3)
#             if(x>=0 and x<height and y>=0 and y<width):
#                 img_resize_array[x,y,k]=img_array[i,j,k]  
    
# for k in range(band):
#     for i in range(img_array.shape[0]):
#         for j in range(img_array.shape[1]):
#             for l in range(1,4):
#                 x=round(i*l)
#                 y=round(j*l)
#                 if(x>=0 and x<height and y>=0 and y<width):
#                     img_resize_array[x,y,k]=img_array[i,j,k]  
plt.figure()
plt.imshow(img_resize_array)
```

# 图像向右向下平移

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

height=img_array.shape[0]*3
width=img_array.shape[1]*3
band=img_array.shape[2]
img_resize_array=np.zeros((height,width,band),dtype=int)
print(img_resize_array.shape)

#平移
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
for k in range(band):
    for i in range(height):
        for j in range(width):
            x=i
            y=j
            img_resize_array[x,y,k]=img_array[i,j,k] 
            x=i+300
            y=j+500
            img_resize_array[x,y,k]=img_array[i,j,k]            
plt.figure()
plt.imshow(img_resize_array)
```

# 图像的正向映射旋转

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

height=img_array.shape[0]*3
width=img_array.shape[1]*3
band=img_array.shape[2]
img_resize_array=np.zeros((height,width,band),dtype=int)
print(img_resize_array.shape)

#正向映射从0点旋转45度
deta=45
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for i in range(height):
        for j in range(width):

            x=round((i)*math.cos(math.radians(deta))-(j)*math.sin(math.radians(deta)))
            y=round((i)*math.sin(math.radians(deta))+(j)*math.cos(math.radians(deta)))

            if(x>=0 and x<height and y>=0 and y<width):
                img_rotate_array[x,y,k]=img_array[i,j,k]            
plt.figure()
plt.imshow(img_rotate_array)

#正向映射从平移点旋转45度
deta=45
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for i in range(height):
        for j in range(width):

            x=round((i)*math.cos(math.radians(deta))-(j)*math.sin(math.radians(deta)))+round(height/2)
            y=round((i)*math.sin(math.radians(deta))+(j)*math.cos(math.radians(deta)))+round(width/2)

            if(x>=0 and x<height and y>=0 and y<width):
                img_rotate_array[x,y,k]=img_array[i,j,k]    
            #x,y已经平移了height/2,width/2
            #再进行平移height
            img_resize_array[x+height,y,k]=img_array[i,j,k]
            
plt.figure()
plt.imshow(img_rotate_array)
plt.figure()
plt.imshow(img_resize_array)
```

# 图像的反向映射旋转

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

#反向映射从0点旋转45度
deta=45
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for x in range(height):
        for y in range(width):

            i=round((x)*math.cos(math.radians(deta))+(y)*math.sin(math.radians(deta)))
            j=round((-x)*math.sin(math.radians(deta))+(y)*math.cos(math.radians(deta)))

            if(i>=0 and i<height and j>=0 and j<width):
                img_rotate_array[x,y,k]=img_array[i,j,k]            
plt.figure()
plt.imshow(img_rotate_array)

#反向映射从平移点旋转45度
#反向映射的平移相当于移动像框
deta=45
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for x in range(height):
        for y in range(width):

            i=round((x-round(height/2))*math.cos(math.radians(deta))+(y-round(height/2))*math.sin(math.radians(deta)))
            j=round(-(x-round(height/2))*math.sin(math.radians(deta))+(y-round(height/2))*math.cos(math.radians(deta)))

            if(i>=0 and i<height and j>=0 and j<width):
                img_rotate_array[x,y,k]=img_array[i,j,k]            
plt.figure()
plt.imshow(img_rotate_array)

#反向映射从平移点旋转45度
#反向映射的平移相当于移动像框
deta=45
height=img_array.shape[0]*3
width=img_array.shape[1]*3
band=img_array.shape[2]

img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for x in range(height):
        for y in range(width):

            i=round((x-round(height/2))*math.cos(math.radians(deta))+(y-round(height/2))*math.sin(math.radians(deta)))
            j=round(-(x-round(height/2))*math.sin(math.radians(deta))+(y-round(height/2))*math.cos(math.radians(deta)))

            if(i>=0 and i<img_array.shape[0] and j>=0 and j<img_array.shape[1]):
                img_rotate_array[x,y,k]=img_array[i,j,k]    

            
plt.figure()
plt.imshow(img_rotate_array)
```

## 只修改其中一个坐标


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

#反向映射从平移点旋转45度
#反向映射的平移相当于移动像框
#只修改其中一个坐标相当于旋转坐标轴后在坐标方向平移后再把坐标轴旋转回来
deta=45
height=img_array.shape[0]
width=img_array.shape[1]
band=img_array.shape[2]
img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for x in range(height):
        for y in range(width):

            i=round((x)*math.cos(math.radians(deta))+(y)*math.sin(math.radians(deta)))-round(200/math.sin(math.pi/4))
            j=round(-(x)*math.sin(math.radians(deta))+(y)*math.cos(math.radians(deta)))-round(100/math.sin(math.pi/4))

            if(i>=0 and i<height and j>=0 and j<width):
                img_rotate_array[x,y,k]=img_array[i,j,k]            
plt.figure()
plt.imshow(img_rotate_array)
```

## 非等比例变换(仿射变换)

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

img = Image.open(r'd:\myjpg\30.jpg')
img_array=np.array(img)
print(img_array.shape)
plt.figure()
plt.imshow(img_array)

#反向映射从平移点旋转45度
#反向映射的平移相当于移动像框
deta=45
height=img_array.shape[0]*5
width=img_array.shape[1]*5
band=img_array.shape[2]

img_rotate_array=np.zeros((height,width,band),dtype=int)
for k in range(band):
    for x in range(height):
        for y in range(width):

            i=round((x/2-500)*math.cos(math.radians(deta))+(y/4)*math.sin(math.radians(deta)))#-round(100/math.sin(math.pi/4))
            j=round(-(x/2-500)*math.sin(math.radians(deta))+(y/4)*math.cos(math.radians(deta)))

            if(i>=0 and i<img_array.shape[0] and j>=0 and j<img_array.shape[1]):
                img_rotate_array[x,y,k]=img_array[i,j,k]    

            
plt.figure()
plt.imshow(img_rotate_array)
```
