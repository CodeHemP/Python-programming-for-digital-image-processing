# 数字图像打开显示

```python
from PIL import Image 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img=Image.open("d:/myjpg/30.jpg")
img1=np.array(img)
print(img1.shape)
plt.figure()
plt.imshow(img1)

img=Image.open("d:/myjpg/31.jpg")
img2=np.array(img)
print(img2.shape)
plt.figure()
plt.imshow(img2)

img=img1//2+img2//2
plt.figure()
plt.imshow(img)
```
